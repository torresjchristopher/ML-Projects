# Fine-tune DistilBERT for Text Classification

This notebook shows a minimal example of fine-tuning a DistilBERT model using the training script.


```python
# 1. Install dependencies (run in your environment, not necessary inside venv if already installed)
!
```

```python
# 2. Prepare dataset (example: small CSV)
import pandas as pd
df = pd.DataFrame({
    'text': ['I love this', 'I hate that', 'This is amazing', 'Terrible experience'],
    'label': [1, 0, 1, 0]
})
df.to_csv('data/dataset.csv', index=False)
df.head()
```

```python
# 3. Run training script (example)
!python train.py --model_name_or_path distilbert-base-uncased --data_file data/dataset.csv --text_col text --label_col label --output_dir outputs/distilbert-sample --num_train_epochs 1 --per_device_train_batch_size 4
```

```python
# 4. Run the server locally (example)
!python serve.py --model_dir outputs/distilbert-sample --port 8080
```
